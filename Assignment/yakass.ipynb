import numpy as np


```python
k = 0.1  # Thermal conductivity
Area = 1  #  area
d = 0.1  # Thickness
T_e = 300  # room tempreture
T_0 = 500  # solid's Initial temperature
t_final = 100  # Final T
dt = 0.01  # Interval
```

```python
t = np.arange(0, t_final + dt, dt)
```

```python
T = np.zeros_like(t)
T[0] = T_0
for i in range(1, len(t)):
    T[i] = T[i-1] - k * Area * dt * (T[i-1] - T_e) / d
```

```python
import matplotlib.pyplot as plt

plt.plot(t, T)
plt.xlabel("Time (s)")
plt.ylabel("Temperature (K)")
plt.show()
```
